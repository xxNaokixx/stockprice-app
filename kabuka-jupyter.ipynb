```python
#まとめ


def get_data(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():

        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period = f"{days}d")
        hist.index = hist.index.strftime("%d %B %Y")
        hist = hist[["Close"]]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = "Name"
        df = pd.concat([df, hist])
    return df
```

```python
days = 2000
tickers = {
    "Apple": "AAPL",
    "Microsoft":"MSFT",
    "META": "META",
    "Tesla": "TSLA",
    "Amazon": "AMZN",
    "Google": "GOOGL", 
    "Netflix": "NFLX"
}

get_data(days, tickers)
```

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8, clip = True)
    .encode(
        x = "Date:T",
        y = alt.Y("Stock Price(USD):Q", stack = None, scale = alt.Scale(domain=[100, 300])),
        color = "Name:N"
    )
)
```
